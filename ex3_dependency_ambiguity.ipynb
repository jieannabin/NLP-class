### Step 0. Install and load
About 30 seconds in Colab.

```python
!pip install -q spacy
!python -m spacy download en_core_web_sm -q

import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
print("model loaded")
```

### Step 1. Parse three sentences
The third one is the classic prepositional-phrase attachment ambiguity. Watch where the arc from *with* points.

```python
sentences = [
    "The analyst flagged the report.",
    "Revenue dropped because the logging rule changed.",
    "I saw the man with the telescope.",
]

for sent in sentences:
    print("\n" + "=" * 60)
    print(sent)
    doc = nlp(sent)
    displacy.render(doc, style="dep", jupyter=True,
                    options={"compact": True, "distance": 110,
                             "color": "#151A2E"})
```

### Step 2. Read the attachment directly
For the telescope sentence, the question is: what does *with* attach to? If its head is *saw*, you used the telescope. If its head is *man*, the man had the telescope. Print the head to see the parser's commitment.

```python
doc = nlp("I saw the man with the telescope.")
for t in doc:
    print(f"{t.text:<12} dep={t.dep_:<10} head={t.head.text}")

with_tok = [t for t in doc if t.text == "with"][0]
print(f"\n'with' attaches to: {with_tok.head.text}  ->  "
      f"the parser committed to ONE of the two readings.")
```
